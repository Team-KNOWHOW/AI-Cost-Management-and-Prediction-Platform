```
import pandas as pd
df = pd.read_excel('20yDataset.xlsx')
print(df.head())
```

```
#결측치 대체
df = df.fillna(0)
```

```
#고정비5개/ 
#각각 비율마다 모든 값에 곱해주는 형식으로 데이터 만들고 나중에 합쳐주기
# x2 = 당기투입 노무비 간접x2,당기투입 노무비 직접 고정비x2, 당기투입 제조경비 간접x2, 당기투입 제조경비 직접 고정비 기타x2, 당기투입 제조경비 직접 고정비 상각비x2
df_1=df[["ic_idlc","ic_dlfc","ic_idohc","ic_ohdfe","ic_ohdfd"]]
```

```
df_1
```

```
count = 0 # 줄의 수를 파악
m_count = 1 # 월 단위 파악
m = 0 # 인덱스
while True:
  if count == 753:
    break
  if (count >=322 and count <=357) or (count >=718 and count <=753):
    next_month = []
    for i in range(len(df_1.columns)):
      #해당 월 금융 위기/코로나로 10%하양조정.
      temp = df_1.values[m][i] * (1 - (0.01 * 10 * m_count/12))
      next_month.append(temp)
    df_1.loc[len(df_1)] = next_month
    count += 1
  else:
    next_month = []
    for i in range(len(df_1.columns)):
      #홀짝수인지 확인 후 1%,2%
      if m_count%2 == 1:
        temp = df_1.values[m][i] * (1 + (0.01 * 1 * m_count/12))
        next_month.append(temp)
      else:
        temp = df_1.values[m][i] * (1 + (0.01 * 2 * m_count/12))
        next_month.append(temp)
    df_1.loc[len(df_1)] = next_month
    count += 1
  #3개 목록 다음에 다음달로 넘어감
  if count%3 == 0:
    m_count += 1
  #12월 다음 조건문 추가
  if m_count >12:
    m_count = 1
  m += 1



# next_month = df.values[1] * 1.001
# df.values.append(3, ignore_index=True)
```

```
#변동비5개/ 
#당기투입 노무비 직접 변동비, 당기투입 제조경비 직접 변동비x1
df_2= df[["ic_dlvc","ic_ohdvc"]]
```

```
count = 0 # 줄의 수를 파악
m_count = 1 # 월 단위 파악
m = 0 # 인덱스
while True:
  if count == 753:
    break
  if (count >=322 and count <=357) or (count >=718 and count <=753):
    next_month = []
    for i in range(len(df_2.columns)):
      #해당 월 금융 위기/코로나로 10%하양조정.
      temp = df_2.values[m][i] * (1 - (0.01 * 10 * m_count/12))
      next_month.append(temp)
    df_2.loc[len(df_2)] = next_month
    count += 1
  else:
    next_month = []
    for i in range(len(df_2.columns)):
      #홀짝수(월)인지 확인 후 동일 3%
      temp = df_2.values[m][i] * (1 + (0.01 * 3 * m_count/12))
      next_month.append(temp)
    df_2.loc[len(df_2)] = next_month
    count += 1
  #3개 목록 다음에 다음달로 넘어감
  if count%3 == 0:
    m_count += 1
  #12월 다음 조건문 추가
  if m_count >12:
    m_count = 1
  m += 1
```

```
#수량,단가, 수량*단가(매출액)
#생산량, 배부기준 생산액 단위당, 배부기준 생산액 적수y
df_3 = df[["proq", "proamt_unit","proamt_acc"]]
```

```
count = 0 # 줄의 수를 파악
m_count = 1 # 월 단위 파악
m = 0 # 인덱스
while True:
  if count == 753:
    break
  if (count >=322 and count <=357) or (count >=718 and count <=753):
    next_month = []
    for i in range(len(df_3.columns)):
      #해당 월 금융 위기/코로나로 10%하양조정.
      temp = df_3.values[m][i] * (1 - (0.01 * 10 * m_count/12))
      next_month.append(temp)
    df_3.loc[len(df_3)] = next_month
    count += 1
  else:
    next_month = []
    for i in range(len(df_3.columns)):
     #홀짝수인지 확인 후 2.5%,3%
      if m_count%2 == 1:
        temp = df_3.values[m][i] * (1 + (0.01 * 2.5 * m_count/12))
        next_month.append(temp)
      else:
        temp = df_3.values[m][i] * (1 + (0.01 * 3 * m_count/12))
        next_month.append(temp)
    df_3.loc[len(df_3)] = next_month
    count += 1
  #3개 목록 다음에 다음달로 넘어감
  if count%3 == 0:
    m_count += 1
  #12월 다음 조건문 추가
  if m_count >12:
    m_count = 1
  m += 1
```

```
df_3
```

```
#반올림
df_1=df_1.round(0).astype(int)
df_2=df_2.round(0).astype(int)
df_3=df_3.round(0).astype(int)
```

```
#분리 연산된 데이터 프레임 합치기
result = pd.concat([df_1,df_2,df_3],axis=1)
result
```

```
#db형태에 맞게 id, version, 
len(result.index)
```

```
#id column inserting code
id = []
for i in range(len(result.index)):
  id.append(i+1)
result.insert(0,'id',id)
```

```
result
```

```
#version_id column inserting code
version_id = []
num = 200001
while True:
  if len(version_id) >= len(result.index):
    break
  if num % 100 == 13:
    num += 100
    num -= 12
  for _ in range(3):
    version_id.append(num)
  num += 1
result.insert(1,'version_id',version_id) 
```

```
#itemplan_id column inserting code
itemplan_id = []
temp = 1
while True:
  if len(itemplan_id) >= len(result.index):
    break
  if temp % 3 == 1:
    itemplan_id.append(3)
  elif temp % 3 == 2:
    itemplan_id.append(2)
  else:
    itemplan_id.append(1)
  temp += 1
result.insert(2,'itemplan_id',itemplan_id)
  
```
