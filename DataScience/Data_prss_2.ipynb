```
import pandas as pd
df = pd.read_excel('20yDataset.xlsx')
print(df.head())
```

```
#결측치 대체
df = df.fillna(0)
```

```

count = 0
m = 0
while True:
  if count == 753:
    break
  next_month = []
  for i in range(len(df.columns)):
    temp = df.values[m][i] * 1.001
    next_month.append(temp)
  df.loc[len(df)] = next_month
  count += 1
  m += 1



# next_month = df.values[1] * 1.001
# df.values.append(3, ignore_index=True)
```

```
#반올림
df=df.round(0).astype(int)
```

```
#db형태에 맞게 id, version, 
len(df.index)
```

```
#id column inserting code
id = []
for i in range(len(df.index)):
  id.append(i+1)
df.insert(0,'id',id)
```

```
#version_id column inserting code
version_id = []
num = 200001
while True:
  if len(version_id) >= len(df.index):
    break
  if num % 100 == 13:
    num += 100
    num -= 12
  for _ in range(3):
    version_id.append(num)
  num += 1
df.insert(1,'version_id',version_id) 
```

```
#itemplan_id column inserting code
itemplan_id = []
temp = 1
while True:
  if len(itemplan_id) >= len(df.index):
    break
  if temp % 3 == 1:
    itemplan_id.append('A-112')
  elif temp % 3 == 2:
    itemplan_id.append('A-108')
  else:
    itemplan_id.append('A-107')
  temp += 1
df.insert(2,'itemplan_id',itemplan_id)
  
```
